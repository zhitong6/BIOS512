````r
df_marathon = read.csv('https://raw.githubusercontent.com/chuckpr/BIOS512/main/in-class-exercises/data/marathon.csv')

df_filtered = df_marathon %>% 
    filter(country %in% c('United States', 'Kenya', 'Ethiopia', 'Japan'))

df_country = df_marathon %>%
    filter(country %in% c('United States', 'Kenya', 'Ethiopia', 'Japan')) %>%
    group_by(country, MF) %>%
    summarize(race.time = mean(race.time), year = mean(year))
````

````r
options(repr.plot.width = 5, repr.plot.height = 4.5, repr.plot.res = 300)

p = ggplot(df_marathon, aes(x = year, y = race.time))

p = p + geom_point(color = 'gray50')

p = p + geom_point(data = df_filtered, aes(color = country))

p = p + geom_text_repel(
    segment.color = 'transparent',
    nudge_y = 0.2,
    data = df_country, 
    aes(label = country, color = country)
)

p = p + facet_wrap(~MF, scales = "free", ncol = 1)

p = p + theme(legend.position = 'none')

p
````


````r
options(repr.plot.width = 5, repr.plot.height = 4.5, repr.plot.res = 300)

p = ggplot(df_marathon, aes(x = year, y = race.time))

p = p + geom_point(color = 'gray50')

p = p + geom_point(data = df_filtered, aes(color = country))

p = p + geom_text_repel(
    segment.color = 'transparent',
    nudge_y = 0.2,
    data = df_country, 
    aes(label = country, color = country)
)

p = p + facet_wrap(~MF, scales = "free", ncol = 1)
        # add title and subtitle; change name for x, y axis
p = p + labs(title = "Men's vs Women's winning race time", 
             subtitle = "running faster over time",
             x = "Year", y = "Winning Time (hour)")

p = p + theme(legend.position = 'none', 
              # background outside the plot
              plot.background=element_rect(fill = "black"),
              # background for the plotting window 
              panel.background=element_rect(fill = "black"),
              # change for background and text color for facet.wrap ("Men's"; "Women's")
              strip.background=element_rect(fill = "black"),
              strip.text = element_text(colour = "white"),
              # title and subtitle for the whole plot 
              plot.title = element_text(colour = "white"),
              plot.subtitle = element_text(colour = "white"),
              # axis title ("Year", "Winning Time (hour)")
              axis.title = element_text(colour = "white"),
              # axis text (numbers on two axis)
              axis.text = element_text(colour = "white"),
              # x and y axis lines
              axis.line = element_line(size = 0.5, colour = "white"),
              # more space around the plot
              plot.margin = unit(c(0.8, 0.8, 0.8, 0.8), "cm"),
              # more distance between two panels
              panel.spacing = unit(1.5, "lines"),
              # title to line up with the left side of "plot" instead of "panel"
              plot.title.position = "plot", 
              # lines within the plot window
              panel.grid.major = element_line(colour = "grey25"),
              panel.grid.minor = element_blank()
             )


p
````
